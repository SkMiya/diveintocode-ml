# ダミーデータの作成

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 【問題１】乱数の作成
* 平均(-3,0)，共分散行列([1.0,0.8],[0.8,1.0])，２次元正規分布に従う乱数500個
* 1.0が分散，0.8が共分散を示す．
* 共分散とは２組の対応するデータがどれほどお互い影響を持ちながら散らばっているか
* 値がそれぞれのデータの分散に近いほど相関性が強くなる
* np.random.multivariate_normal(平均,共分散行列,数)

```python
# 平均
m = np.array([-3,0])
# 共分散行列
c = np.array([[1.0,0.8],[0.8,1.0]])
# シードの固定
np.random.seed(0)
# 乱数作成
x=np.random.multivariate_normal(m,c,500)
# shapeの確認
x.shape
```

# 【問題２】散布図による可視化

```python
plt.scatter(x[:,0],x[:,1])
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('scatter')
```

# 【問題３】ヒストグラムによる可視化

```python
plt.figure(figsize=(8,8))
# １列目
plt.subplot(2,1,1)
plt.xlim(-6,3)
plt.hist(x[:,0],bins=50)
plt.title('histgram of x1')
plt.xlabel('x1')
plt.ylabel('frequency')
# ２列目
plt.subplot(2,1,2)
plt.hist(x[:,1],bins=50)
plt.xlim(-6,3)
plt.title('histgram of x2')
plt.xlabel('x2')
plt.ylabel('frequency')
plt.show()
```

# 【問題４】データの追加
* 平均(0,-3)，共分散行列([1.0,0.8],[0.8,1.0])の２次元正規分布に従う乱数を500個
* 上記と１の散布図を可視化．凡例は０，１とする

```python
# 平均
m2 = np.array([0,-3])
# 共分散行列
c2 = np.array([[1.0,0.8],[0.8,1.0]])
# シード固定
np.random.seed(0)
# 乱数作成
x2 = np.random.multivariate_normal(m2,c2,500)

# 可視化
plt.scatter(x[:,0],x[:,1],label='0')
plt.scatter(x2[:,0],x2[:,1],label='1')
plt.legend()
plt.title('scatter')
plt.xlabel('x1')
plt.ylabel('x2')
```

# 【問題５】データの結合
問題１と問題４のndarrayを結合

### np.concatenate([],axis=0)…axisを１にすると横方向

```python
concatenate_x = np.concatenate([x,x2],axis=0)
concatenate_x.shape
```

### np.vstack(())…hstackは横方向

```python
vstack_x = np.vstack((x,x2))
vstack_x.shape
```

# 【問題６】ラベル付け
* 結合したデータに問題１，４どちらで作成したものなのか示すラベルを付ける
* ０と１を含むラベル列を新たに追加

```python
# 問題１のデータには０を追加
x_label = np.insert(x,2,0,axis=1)
# 問題４のデータには１を追加
x2_label = np.insert(x2,2,1,axis=1)
print('x1:\n',x_label)
print('\nx2:\n',x2_label)
```

```python
# 上のデータをくっつける
x_add_label = np.concatenate([x_label,x2_label],axis=0)
print('newshape:',x_add_label.shape)
print('\n',x_add_label)
```
